## (4) Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points) 

```python
df=df.astype({'GoalsFor': 'int64','GoalsAgainst': 'int64'})


def Goal_diff_count(df,case):
    if(case):
        df['Goal Difference'] = df['GoalsFor']-df['GoalsAgainst']
        return df
    else:
        goal_max= df.nlargest(1,['Goal Difference']).loc[:,['Team','Goal Difference']]
        print("\n Maximum Goal Difference ==> \n", goal_max)
        goal_min= df.nsmallest(1,['Goal Difference']).loc[:,['Team','Goal Difference']]
        print("\n Minimum Goal Difference ==> \n", goal_min)
        
        
x=Goal_diff_count(df,True)
print(x.loc[:,['Team','Goal Difference']])
Goal_diff_count(df,False)
```

## (5)  Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
#conversion of type
df=df.astype({'GamesWon': 'int64','GamesPlayed': 'int64'})

Percentage_of_Winning =((df['GamesWon']) / (df['GamesPlayed']))*100


df['Winning Percent']=Percentage_of_Winning
df['Winning Percent']=df['Winning Percent'].fillna(0)
df
```
