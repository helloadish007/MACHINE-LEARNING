```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]#week days
y=[20,30,37,32,40,34,56]#temp corresponding to the week days
```

```python
plt.plot(x,y,color='red',linewidth=5)
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

```python
plt.plot(x,y,color='red',linewidth=5,linestyle='dashdot')
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,'g+')#g-greencolor---+-symbol
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,'g*')#g-greencolor---*-symbol
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,'rD--')#D-diamond
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,'rD--',markersize=10)#D-diamond
```

```python
#setting x and ylabel and title
plt.xlabel('week')
plt.ylabel('temperature')
plt.title('temp_week plot')
plt.plot(x,y,'rD--',alpha=0.4,markersize=10)#alpha(0-1range)--controls transparency
```

```python

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('temp')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best')
plt.grid()
```

```python

import numpy as np
x = np.linspace(-5,5,100)
y = 2*x+1
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='red')
plt.ylabel('y', color='green')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
