```python
import pandas as pd
df=pd.read_csv('homep.txt')
df.columns=['town','area','price']
df.to_csv('homeprice_text.csv')
df
```

### how to handle text data in a model?

use label encoding/integer encoding..to convert textual data to numeric data.

### problem:

Assume that you assigned like:

Monroe township = 1 , west windsor = 2 ,robinsville = 3

but the model, will assume some kind of orders between these numeric values like:

Monroe township < west windsor < robinsville

            or
Monroe township + west windsor = robinsville 

### NOMINAL: categaries without any kind of ordering.eg: male,female

### ORDINAL: categories with some kind of numerical ordering.eg:high,medium,low

### solution :one hot encoding

create a new column for each of your categories and assign binary values(1/0).

these extra columns are called DUMMY VARIABLES.

```python
dum=pd.get_dummies(df.town)
dum
```

```python
join=pd.concat([df,dum],axis='columns')#join original dataframe with the dummies
join
```

### Dummy Variable Trap

whenever,one variable can be derived from rest of the variables,these variables are said to be multicollinear.

multicollinearity confuses model to assume some sort of relation between the dummy variables created

### solution:

drop one dummy variable column

```python
df1=join.drop(['town','monroe township'],axis='columns')
df1
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=df1.drop('price',axis='columns')
x
```

```python
y=df1.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2600,1,0]])#predicting price of the house at robinsville 
```

```python
model.predict([[2600,0,0]])#predicting price of the house at monroe township
```

```python
model.score(x,y)#finding accuracy of the model
```
