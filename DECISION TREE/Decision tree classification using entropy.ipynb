```python
x=df.iloc[:,:-1].values#here iloc selects all column except last # '.values'-->convert into numpy array.
y=df.iloc[:,2].values
```

###  If you don't specify the random_state in your code, then every time you run(execute) your code a new random value is generated and the train and test datasets would have different values each time.

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

# Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
xtrain
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(xtrain,ytrain)
```

```python
classifier.predict(sc.transform([[49,36000]]))
```

```python
ypred=classifier.predict(xtest)
print(np.concatenate((ypred.reshape(len(ypred),1),ytest.reshape(len(ytest),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(ytest,ypred)
print(cm)
accuracy_score(ytest,ypred)
```
