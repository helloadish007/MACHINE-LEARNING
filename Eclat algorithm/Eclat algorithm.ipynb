### Eclat is an algorithm for discovering frequent itemsets in a transaction database. It was proposed by Zaki (2001). Contrarily to algorithms such as Apriori, Eclat uses a depth-first search for discovering frequent itemsets instead of a breath-first search.

#### What is the input of the Eclat algorithm?

#### The input is a transaction database.A transaction database is a set of transactions. Each transaction is a set of items.

#### What is the output of the Eclat algorithm?

#### Eclat is an algorithm for discovering itemsets (group of items) occurring frequently in a transaction database (frequent itemsets). A frequent itemset is an itemset appearing in at least minsup transactions from the transaction database, where minsup is a parameter given by the user.

#### Advantages over Apriori algorithm:-

- Memory Requirements: Since the ECLAT algorithm uses a Depth-First Search approach, it uses less memory than Apriori algorithm.
- Speed: The ECLAT algorithm is typically faster than the Apriori algorithm.
- Number of Computations: The ECLAT algorithm does not involve the repeated scanning of the data to compute the individual support values.

## In coding , we use the following steps:

Step 1: Set a minimum support.



Step 2: Take all the subsets in a transaction having higher support than minimum support.



Step 3: Sort these subsets by decreasing support.



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#apriori function in apyori module will not take panda's dataframe as input, instead it takes input as
# a list of transactions.
df = pd.read_csv('/home/adish/Desktop/Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
df
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
results
```

```python
#results converted into a dataframe
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
rdf = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

```python
rdf.nlargest(n = 10, columns = 'Support')#based on min.support specified by user..not based on confidence and lift values
```
