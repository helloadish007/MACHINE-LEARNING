```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/home/adish/Desktop/Social_Network_Ads.csv')
x= df.iloc[:, :-1].values
y= df.iloc[:, -1].values
df
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
msc =MinMaxScaler()
xtrain = msc.fit_transform(xtrain)
xtest = msc.transform(xtest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(xtrain,ytrain)
```

```python
classifier.predict(msc.transform([[20,20000]]))
```

```python
y_pred = classifier.predict(xtest)
print(np.concatenate((y_pred.reshape(len(y_pred),1), ytest.reshape(len(ytest),1)),1))
```

```python
classifier.score(xtest,ytest)#gives the accuracy of the KNN model
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(ytest, y_pred)
print(cm)
#(4+3)=7 incorrect predictions, 64 correct prediction of the class 0 and 29 correct prediction of the class 1 
```
