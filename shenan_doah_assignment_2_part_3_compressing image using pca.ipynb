6. An image is a matrix of the gray scale. What if we pick the principal components for the matrix? The image is compressed! To understand the data compression by PCA:

* Load and plot a Franklin University logo:

```python
from skimage import data, io, color
from pylab import *
img = io.imread("https://github.com/franklin-univ-data-science/data/blob/master/images/Franklin_logo.jpg?raw=true", as_gray=True)
xlabel('Original Image')
io.imshow(img)
```

 What is the dimension of the logo?

```python
img.shape
```

```python
import pylab as plt
import numpy as np


print(img.shape)
plt.axis('off') 
plt.imshow(img)
plt.show()
```

```python

img_reshaped = np.reshape(img, (400, 400))
print(img_reshaped.shape)
```

```python

from sklearn.decomposition import PCA

pca = PCA(n_components=12)
pca.fit(img_reshaped)

img_transformed = pca.transform(img_reshaped)
print(img_transformed.shape)
```

```python
img_inverse = pca.inverse_transform(img_transformed)
print(img_inverse.shape)

plt.imshow(img_inverse)
plt.show()

#the total number of principal components which make the logo still visible = 12 or can be 13 also
```

### when the total number of principal components which make the logo still visible =  13 

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=13)
pca.fit(img_reshaped)

img_transformed = pca.transform(img_reshaped)
print(img_transformed.shape)
```

```python
img_inverse = pca.inverse_transform(img_transformed)
print(img_inverse.shape)

plt.imshow(img_inverse)
plt.show()

#the total number of principal components which make the logo still visible = 12 or can be 13 also
```
