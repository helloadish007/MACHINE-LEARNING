 If we try to use a simple linear regression in the above graph then the linear regression line won’t fit very well. It is very difficult to fit a linear regression line in the above graph with a low value of error. Hence we can try to use the polynomial regression to fit a polynomial line so that we can achieve a minimum error or minimum cost function. The equation of the polynomial regression for the above graph data would be:
 
y = θo + θ₁x₁ + θ₂ x₁²

the general equation of a polynomial regression is:

Y=θo + θ₁X + θ₂X² + … + θₘXᵐ + residual error

#### Polynomial Regression is a form of linear regression in which the relationship between the independent variable x and dependent variable y is modeled as an nth degree polynomial

## Uses of Polynomial Regression:

#### These are basically used to define or describe non-linear phenomenon such as:

- Growth rate of tissues.
- Progression of disease epidemics
- Distribution of carbon isotopes in lake sediments

## Note: A Polynomial Regression algorithm is also called Polynomial Linear Regression because it does not depend on the variables, instead, it depends on the coefficients, which are arranged in a linear fashion.

### Regression overview

Simple Linear Regression equation:         y = b0+b1x         .........(a)

Multiple Linear Regression equation:         y= b0+b1x+ b2x2+ b3x3+....+ bnxn         .........(b)

Polynomial Regression equation:         y= b0+b1x + b2x2+ b3x3+....+ bnxn         ..........(c)

## scenario:

- Let's assume that we are conducting a recruitment process and we need to hire some professionals.
    let's suppose while interviewing a person we just asked him about his expected salary and he may reply with a genuine value based on his previous salary or there's a chance that he might bluff .Inorder to avoid this situation and learn about the range of salary that a person could obtain sitting at a certain level  can  be find out using our polynomial regression technique.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/home/adish/Desktop/Position_Salaries.csv')
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
df
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
###multilinear regression
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)#matrix of powered features-->x,x^2,x^3 etc
poly_reg1 = LinearRegression()
poly_reg1.fit(x_poly, y)
#y=b0+b1*(position_level)+b2*(position_level)^2+b3*(position_level)^3+b4*(position_level)^4
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, reg.predict(x_poly), color = 'green')
plt.title('Truth or Bluff using Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Displaying Polynomial regression plots

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_reg1.predict(x_poly), color = 'green')
plt.title('Truth or Bluff using polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
reg.predict([[6.5]])#predict salary at level 6.5===> at level 6 ,he has a salary of 1,50,000
```
