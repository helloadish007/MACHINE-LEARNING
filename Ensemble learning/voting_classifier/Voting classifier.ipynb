```python
## importing models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB 
from sklearn.svm import SVC

from sklearn.ensemble import VotingClassifier

import pandas as  pd

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


import warnings
warnings.filterwarnings("ignore")
```

```python
import os 
os.listdir()
```

```python
df=pd.read_csv('/home/adish/Downloads/heart.csv')
df
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# test and Train Split

xtrain,xtest,ytrain,ytest = train_test_split( x,y,test_size=.25,random_state=4)
```

```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC()
nbg_clf = GaussianNB()
```

```python
for clf in (log_clf, nbg_clf, rnd_clf, svm_clf):
    clf.fit(xtrain, ytrain)
    
    y_pred = clf.predict(xtest)
    
    print(clf.__class__.__name__, accuracy_score(ytest, y_pred))
```

```python
voting_clf = VotingClassifier(estimators=[('lr', log_clf),('nb',nbg_clf), ('rf', rnd_clf), ('svc', svm_clf)],voting='hard')

for clf in (log_clf, nbg_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(xtrain, ytrain)
    y_pred = clf.predict(xtest)
    print(clf.__class__.__name__, accuracy_score(ytest, y_pred))
```
